```python
import covsirphy as cs
import matplotlib.pyplot as plt
from datetime import timedelta
```

```python
country = "Lithuania"
scenario_1 = cs.ODEScenario.auto_build(geo=country, model=cs.SIRFModel)
```

```python

scenario_1.summary()

```

```python

scenario_1.simulate(name="Baseline")
```

```python

scenario_1.track()
```

```python
scenario_1.compare_param("Rt")
scenario_1.compare_param("theta")
scenario_1.compare_param("kappa")
scenario_1.compare_param("rho")
scenario_1.compare_param("sigma")
scenario_1.compare_param("tau")
```

```python

scenario_1.simulate(name="Baseline")
```

## **Передбачення**

```python
future_start_date = scenario_1.simulate(display=False).index.max() + timedelta(days=1)
future_start_date
```

# **На 7 днів**


```python
scenario_1.build_with_template(name="Predicted", template="Baseline");

scenario_1.predict(days=7, name="Predicted")
```

```python
df = scenario_1.append().summary()
df.loc[df["Start"] >= future_start_date]
```

```python

scenario_1.simulate(name="Predicted")
```
